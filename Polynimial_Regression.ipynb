```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

```

```python
dataset= pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:,-1].values
X
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X,y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(X)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly,y)
```

```python
plt.scatter(X,y, color='red')
plt.plot(X, lin_reg.predict(X), color='blue')
plt.title('Truth or Bluff (Linear Regresison)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
plt.scatter(X,y, color='red')
plt.plot(X, lin_reg_2.predict(X_poly), color='blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
f, ax = plt.subplots(1,2, figsize=(15,5))

ax[0].scatter(X,y, color='red')
ax[0].plot(X, lin_reg.predict(X), color='blue')
ax[0].set_title('Truth or Bluff (Linear Regresison)')
ax[0].set_xlabel('Position Level')
ax[0].set_ylabel('Salary')

ax[1].scatter(X,y, color='red')
ax[1].plot(X, lin_reg_2.predict(X_poly), color='blue')
ax[1].set_title('Truth or Bluff (Polynomial Regression)')
ax[1].set_xlabel('Position Level')
ax[1].set_ylabel('Salary')
```

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid),1))
X_poly = poly_reg.fit_transform(X_grid)
plt.scatter(X,y, color='red')
plt.plot(X_grid, lin_reg_2.predict(X_poly), color='blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```
