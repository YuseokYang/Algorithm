```python
import pandas as pd
filename = 'iris.csv'
iris = pd.read_csv(filename)
iris.head()
```

```python
count = iris['Species'].value_counts()
count
```

```python
iris.isnull().sum()
```

```python
dup = iris.duplicated()
```

```python
iris[dup]
```

```python
iris = iris.drop_duplicates()
```

```python
iris.duplicated().sum()
```

```python
iris[(iris['SepalLength']==5.8) & (iris['PetalWidth']==1.9)
]
```

```python
iris.groupby('Species').sum()
```

```python
iris.groupby('Species').mean()
```

```python
import matplotlib.pyplot as plt
iris[['SepalLength', 'Species']][:10].plot.bar(x='Species', y='SepalLength', title='Sepal Length')
```

```python
df2 = iris.groupby('Species').mean()
```

```python
df2['SepalLength'].plot.bar(rot=0)
plt.title('Sepal Length')
plt.ylabel('MeanLength')
plt.show()
```

```python
iris.plot.hist(bins=30, title='Histogram of iris')
plt.xlabel('values')
plt.ylabel('frequency')
plt.show()
```
