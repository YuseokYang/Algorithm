```python
import matplotlib.pyplot as plt
```

```python
plt.title('Line Graph')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
xdata = [2,3,4,5]
ydata=[1,3,5,7]
plt.plot(xdata, ydata)
```

```python
plt.title('Legend')
data1 = [2,3,4,5]
data2 = [8,6,4,2]
x = [1,2,3,4]
plt.plot(x, data1, label='asc')
plt.plot(x, data2, label='desc')
plt.legend(loc='upper left')
plt.show()
```

```python
plt.plot(xdata,ydata, color='red', label='red graph')
plt.legend()
plt.show()
```

```python
plt.plot(x,data1, color='red', label='dashed', linestyle='--')
plt.plot(x,data2, color='green', label='dotted', linestyle=':')
plt.legend()
plt.show()
```

```python
plt.title('X, Y range')
xdata = [5,10,15,20]
ydata = [1,3,5,7]

plt.plot(xdata,ydata, color='blue', linestyle='--', marker='o', markersize=10)
plt.xlim(0,25)
plt.ylim(-2,10)
plt.show()
```

```python
data = pd.read_excel('fine_dust.xlsx')
data
```

```python
data.info()
```

```python
data.describe()
```

```python
data[[2018,'area']].plot.bar(x='area', y=2018, rot=60, title='2018 Fine dust')
```

```python
data.groupby('area').mean().loc['Busan'].plot.bar()
plt.title('Busan Fine dust')
plt.xlabel('year')
plt.ylabel('value')
```

```python
mask_area=data.groupby('area').mean()
mask_area.loc[:, 2018]
```

```python
plt.figure(figsize=(14,6))
for year in range(2016,2020):
  chartdata = mask_area.loc[:,year]
  plt.plot(chartdata, label=str(year), marker='o')


plt.xlabel('area')
plt.ylabel('micrometer')
plt.title('2016~2019 fine dust line graph')
plt.legend()
plt.grid()
plt.show()
```

```python
data2017 = data[[2017, 'area']]
data2017.plot.bar(x='area', y=2017, rot=60)
plt.title('2017 Fine dust')
plt.ylabel('value')
plt.grid()
```
